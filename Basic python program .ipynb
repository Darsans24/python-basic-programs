print numbers in ascending order

```python
def getSortedNumber(n):

	number = str(n)

	number = ''.join(sorted(number))

	number = int(number)

	return number

n = 193202042

print(getSortedNumber(n))
```

printing finonacci number

```python
n = int(input("enter a number"))
a = 0
b = 1
if(n < 0):
    print(a)
else:
    print(a,b,end=" ")

for i in range(2,n):
    c = a + b
    print(c,end=" ")
    a = b
    b = c
```

printing factorial number

```python
def factorial(n):

    return 1 if (n==1 or n==0) else n * factorial(n - 1)

num = 5
print("Factorial of",num,"is",factorial(num))
```

printing prime numbers

```python
num = 29


f = False

if num == 1:
    print(num, "is not a prime number")
elif num > 1:

    for i in range(2, num):
        if (num % i) == 0:

            f = True

            break


if f:
    print(num, "is not a prime number")
else:
    print(num, "is a prime number")
```

printing armstrong numbers

```python
num = int(input("Enter a number:"))
sum = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum = sum + digit **3
    temp = temp // 10

if num == sum:
    print(num,"amstrong")
else:
    print(num,"not amstrong")
```

reverse a number in python

```python
num = int(input())
rev = 0

while num != 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed Number: " + str(rev))
```
